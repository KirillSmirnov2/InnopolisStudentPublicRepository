```python
# @title Определим множества {Вина старого света} и {Вина нового света}
All_World=set()
All_World=All_World.union(Country_df0_1_red).union(Country_df0_2_ros).union(Country_df0_3_spk).union(Country_df0_4_wht)
print(len(All_World),'стран всего в датасете: ',All_World)
print('--------------------------------')
Old_World_Classic={'Portugal','Italy', 'Greece','Germany','Spain','France'}
Old_World_Other={'Slovakia', 'Bulgaria', 'Czech Republic','Romania','Hungary','Croatia', 'Slovenia','United Kingdom','Moldova','Moldova', 'Israel','Lebanon','Luxembourg','Georgia','Turkey','Switzerland'}
New_World=All_World - Old_World_Other - Old_World_Classic
print(len(Old_World_Classic),'стран классического старого света всего в датасете: ',Old_World_Classic)
print(len(Old_World_Other),'стран остального старого света всего в датасете: ',Old_World_Other)
print(len(New_World),'стран старого света всего в датасете: ',New_World)
#Country_df0_1_red
#Country_df0_2_ros
#Country_df0_3_spk
#Country_df0_4_wht
```

```python
# @title Объединение concat в один dataframe из 4 таблиц
df3_1_red=df2_1_red.copy(deep=True)
df3_2_ros=df2_2_ros.copy(deep=True)
df3_3_spk=df2_3_spk.copy(deep=True)
df3_4_wht=df2_4_wht.copy(deep=True)
n1red=len(df3_1_red['Name'])
n2ros=len(df3_2_ros['Name'])
n3spk=len(df3_3_spk['Name'])
n4wht=len(df3_4_wht['Name'])
ST1=[]
for i in range(n1red):
  ST1.append('red')
df3_1_red['WineType'] = ST1
ST1=[]
for i in range(n2ros):
  ST1.append('ros')
df3_2_ros['WineType'] = ST1
ST1=[]
for i in range(n3spk):
  ST1.append('spk')
df3_3_spk['WineType'] = ST1
ST1=[]
for i in range(n4wht):
  ST1.append('wht')
df3_4_wht['WineType'] = ST1
ST1=[]


df4_combine = pd.concat([df3_1_red, df3_2_ros, df3_3_spk, df3_4_wht], axis=0)
print(df4_combine)
# Using 'Address' as the column name
# and equating it to the list
del n1red, n2ros, n3spk, n4wht , ST1, df3_1_red, df3_2_ros, df3_3_spk, df3_4_wht
```

```python
# @title Определение уникальных значений
n_rows_df4=len(df4_combine)
Country_df4=set(df4_combine['Country'].unique())
Region_df4=set(df4_combine['Region'].unique())
Winery_df4=set(df4_combine['Winery'].unique())
Year_df4=set(df4_combine['Year'].unique())

print('Country_df4_combine : unique count ', len(Country_df4),Country_df4)
print('Region_df4_combine  : unique count ', len(Region_df4),Region_df4)
print('Winery_df4_combine  : unique count ', len(Winery_df4),Winery_df4)
print('Year_df4_combine    : unique count ', len(Year_df4),Year_df4)
```

```python
# @title
L_Old_World_Classic, L_Old_World_Other, L_New_World, ST1=[], [], [], []
for i in range(n_rows_df4):
  s=df4_combine['Country'].iloc[i]
  if s in Old_World_Classic:
    L_Old_World_Classic.append(i)
    ST1.append('Old_World_Classic')
  else:
    if s in Old_World_Other:
      L_Old_World_Other.append(i)
      ST1.append('Old_World_Other')
    else:
      L_New_World.append(i)
      ST1.append('New_World')
df4_combine['Origin'] = ST1
#print(L_Old_World_Classic)
#print(L_Old_World_Other)
#print(L_New_World)
#df3.loc[(df3.isna())['bmi'] == False].index


df4_Old_World_Classic=df4_combine.iloc[L_Old_World_Classic]
df4_Old_World_Other=df4_combine.iloc[L_Old_World_Other]
df4_New_World=df4_combine.iloc[L_New_World]
df5=pd.concat([df4_Old_World_Classic, df4_Old_World_Other, df4_New_World],  keys = ['Old_World_Classic', 'Old_World_Other','L_New_World'])
#Country_df4_control=set(df4_New_World['Country'].unique())

print(df5)
#df4_grby_Worlds=df4_combine.groupby(['Brand','VehicleType'])['Price'].mean().head(20)
```

```python
# @title Группировка
grouped = df5.groupby(['Origin'])['Rating'].mean()
print(grouped)
grouped = df5.groupby(['Origin'])['NumberOfRatings'].sum()
print(grouped)
```

Обобщенный рейтинг вин классического старого света немногим выше, количество оценок по винам стран классического старого света небольшое

```python
grouped = df5.groupby(['Origin','WineType'])['Price'].mean()
print(grouped)
```
