```python
def Count_Members(fls,AL_G, STR_PRT, NM):
  T0 = time.time()
  print('==============================================')
  Kd=0
  Kd1=0
  print('[', end='')
  for j in range(len(AL_G)):
    if fls == 1:
      NM=NM+df1['members'].loc[df1['anime_id'] == AL_G[j]].values
    elif fls == 2:
      NM=NM+len(dfT.loc[dfT['anime_id'] == AL_G[j]])
    else:
      pass
    Kd1=j/len(AL_G)*10 // 1
    if Kd1>Kd:
      Kd = Kd1
      print('-', end='')
  T1 = time.time()
  print('-]')
  print(STR_PRT,' - ОК; ProcessTime: ', (T1-T0), ' sec')

```
