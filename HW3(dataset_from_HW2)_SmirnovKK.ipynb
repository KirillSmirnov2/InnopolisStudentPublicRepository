Определим списки [anime_id] по каждому жанру:  [Shounen]; [Hentai]; [Mecha]
Определим число участников сообществ [members] по [anime_id] каждому жанру по данным из [anime.csv]

определим число участников сообществ по оценкам [rating]  от [user_id] из [rating.csv]   по определенным [anime_id] (по данным из [anime.csv])


```python
def Count_Members(fls,AL_G, STR_PRT, NM):
  T0 = time.time()
  Kd=0
  Kd1=0
  print('[', end='')
  for j in range(len(AL_G)):
    if fls == 1:
      NM=NM+df1['members'].loc[df1['anime_id'] == AL_G[j]].values
    elif fls == 2:
      NM=NM+len(dfT.loc[dfT['anime_id'] == AL_G[j]])
    else:
      pass
    Kd1=j/len(AL_G)*10 // 1
    if Kd1>Kd:
      Kd = Kd1
      print('-', end='')
  T1 = time.time()
  print('-]')
  print(STR_PRT,' - ОК; ProcessTime: ', (T1-T0), ' sec')
  SR=NM
  return SR
```

```python
AnimeIdList_genre_Shounen=[]
AnimeIdList_genre_Hentai=[]
AnimeIdList_genre_Mecha=[]
for i in range(n_rows):
  S1=[]
  if type(df1['genre'].loc[i]) == str:
    S1=df1['genre'].loc[i].split(',')
    P=[]
    for j in S1:
      P=j.replace(' ','')
      if P=='Shounen':
        AnimeIdList_genre_Shounen.append(df1['anime_id'].loc[i])
      else:
        if P=='Hentai':
          AnimeIdList_genre_Hentai.append(df1['anime_id'].loc[i])
        else:
          if P=='Mecha':
            AnimeIdList_genre_Mecha.append(df1['anime_id'].loc[i])
          else:
            pass
  else:
    pass

print('==============================================')
print('количество аниме жанра [Shounen]:   ',len(AnimeIdList_genre_Shounen))
print('количество аниме жанра [Hentai]:   ',len(AnimeIdList_genre_Hentai))
print('количество аниме жанра [Mecha]:   ',len(AnimeIdList_genre_Mecha))
print('==============================================')
NM_v1_Shounen, NM_v2_Shounen, NM_v1_Hentai, NM_v2_Hentai, NM_v1_Mecha, NM_v2_Mecha=0, 0, 0, 0, 0, 0

#efev=df1['members'].loc[df1['anime_id'] == 820]
#print(efev.values)
  #return

NM_v1_Shounen = Count_Members(1,AnimeIdList_genre_Shounen,'NM_v1_Shounen',NM_v1_Shounen)
NM_v1_Hentai=Count_Members(1,AnimeIdList_genre_Hentai,'NM_v1_Hentai',NM_v1_Hentai)
NM_v1_Mecha=Count_Members(1,AnimeIdList_genre_Mecha,'NM_v1_Mecha',NM_v1_Mecha)
print('==============================================')
NM_v2_Shounen=Count_Members(2,AnimeIdList_genre_Shounen,'NM_v2_Shounen',NM_v2_Shounen)
NM_v2_Hentai=Count_Members(2,AnimeIdList_genre_Hentai,'NM_v2_Hentai',NM_v2_Hentai)
NM_v2_Mecha=Count_Members(2,AnimeIdList_genre_Mecha,'NM_v2_Mecha',NM_v2_Mecha)
print('==============================================')

print('================== по файлу anime.csv =======================')
print('количество участников сообществ [members] по [anime_id] жанра [Shounen]:   ', NM_v1_Shounen)
print('количество участников сообществ [members] по [anime_id] жанра [Hentai]:   ', NM_v1_Hentai)
print('количество участников сообществ [members] по [anime_id] жанра [Mecha]:   ', NM_v1_Mecha)
print('==============================================')
print('================== по файлу rating.csv =======================')
print('количество участников сообществ [members] по [anime_id] жанра [Shounen]:   ', NM_v2_Shounen)
print('количество участников сообществ [members] по [anime_id] жанра [Hentai]:   ', NM_v2_Hentai)
print('количество участников сообществ [members] по [anime_id] жанра [Mecha]:   ', NM_v2_Mecha)
print('==============================================')

```
